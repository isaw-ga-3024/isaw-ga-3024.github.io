```python
import os
import re
import bleach
from cltk.corpus.greek.beta_to_unicode import Replacer
from bs4 import BeautifulSoup
from cltk.corpus.greek.tlgu import TLGU
from cltk.corpus.utils.formatter import assemble_tlg_author_filepaths
from cltk.corpus.utils.importer import CorpusImporter
corpus_importer=CorpusImporter('greek')
corpus_importer.import_corpus('greek_models_cltk')
from cltk.stem.lemma import LemmaReplacer
from cltk.corpus.utils.importer import CorpusImporter
corpus_importer=CorpusImporter('greek')
corpus_importer.import_corpus('greek_models_cltk')
from cltk.stem.lemma import LemmaReplacer
from cltk.corpus.utils.formatter import tlg_plaintext_cleanup
lemmatizer=LemmaReplacer('greek')
import sys
from nltk.tokenize.punkt import PunktLanguageVars
from cltk.stop.greek.stops import STOPS_LIST
from collections import defaultdict
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
from gensim import corpora, models, similarities
import gensim
import numpy as np
import pandas as pd
import sqlite3

#the following code for importing the corpus and converting it from beta code and .xml to unicode and .txt
#is from the cltk. It was found in the corpora directory.

root = '/Users/wimsey'
perseus_root = root + '/cltk_data/greek/text/greek_text_perseus/'
authors = os.listdir(perseus_root)

def super_organize(text):
    p = PunktLanguageVars()
    text=tlg_plaintext_cleanup(text, rm_punctuation=True, rm_periods=False)
    text2=lemmatizer.lemmatize(text)
    tokens = p.word_tokenize(' '.join(text2))
    organize_text=[w for w in tokens if not w in STOPS_LIST]
    return organize_text

for author in authors:
    if not '.' in author:
        texts = os.listdir(perseus_root + author + '/opensource')
        for text in texts:
            text_match = re.match(r'soph.aj_gk.xml', text)
            if text_match:
                gk_file = text_match.group()
                txt_file = perseus_root + author + '/opensource/' + gk_file
                with open(txt_file) as gk:
                    xml=gk.read()
                    soup=BeautifulSoup(xml, 'xml')
                    completeText=''
                    for lines in soup.find_all('l'):
                        completeText += str(lines.string)
                    beta_code_text = bleach.clean(completeText, strip=True).upper()
                    a_replacer = Replacer()
                    unicode_converted = a_replacer.beta_code(beta_code_text)
                    clean_greek_list=super_organize(unicode_converted)  
                    unicode_directory = root + '/cltk_data/greek/text/perseus_unicode/'
                    unic_file_exist = os.path.isdir(unicode_directory)
                    if unic_file_exist is True:
                        pass
                    else:
                        os.mkdir(unicode_directory)
                    author_path = unicode_directory + author
                    author_path_pres = os.path.isdir(author_path)
                    if author_path_pres is True:
                        pass
                    else:
                        os.mkdir(author_path)
                    gk_file_txt = os.path.splitext(gk_file)[0] + '.txt'
                    uni_write = author_path + '/' + gk_file_txt
                    with open(uni_write, 'w') as uni_write:
                        frequency = defaultdict(int)
                        for token in clean_greek_list:
                            frequency[token] += 1
                        print(frequency)
                        
                        #db=sqlite3.connect(':memory:')
                        #def init_db(cur):
                            #cur.execute('''CREATE TABLE wordcloud(
                               # Word TEXT,
                                #Frequency INTEGER)'''
                                )
                        #def populate_db(cur, csv_fp):
                            #cur.executemany('''
                            #INSERT INTO wordcloud (Word, Frequency)
                            #VALUES (?,?)''', rdr)
                        
                    
                        
                        #dictionary = corpora.Dictionary(frequency)
                        #dictionary.save('/tmp/soph.dict')  # store the dictionary, for future reference
                        # print(dictionary)
                        #new_vec = dictionary.doc2bow(frequency)
                        #print(new_vec)
                        
                        #clean_greek_string=' '.join(clean_greek_list)
                        #clean_greek_string.join(clean_greek_list)
                        #print(clean_greek_string)
                        #uni_write.write(clean_greek_string)
```
