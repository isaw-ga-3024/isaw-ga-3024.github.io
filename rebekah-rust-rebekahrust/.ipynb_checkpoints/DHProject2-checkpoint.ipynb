# Back to converting the files

```python
#the above code for deleting the header has been incorporated into the code below.
#Note that the following code will take about ten minutes to execute. My mac's activity monitor was useful in monitoring
#the progress. The CPU should be at 100% as the code is being executed.
#Also note that before executing this code, you need to make one modifcation. In the "greek_text_perseus" file there is a file
#named "json". I don't know why this file is here. Rename it to "json.file" in order to avoid an error.
import os
import re
import bleach
from cltk.corpus.greek.beta_to_unicode import Replacer
from bs4 import BeautifulSoup

root = '/Users/wimsey'
perseus_root = root + '/cltk_data/greek/text/greek_text_perseus/'
authors = os.listdir(perseus_root)

for author in authors:
    if not '.' in author:
        texts = os.listdir(perseus_root + author + '/opensource')
        for text in texts:
            text_match = re.match(r'.*_gk.xml', text)
            if text_match:
                gk_file = text_match.group()
                txt_file = perseus_root + author + '/opensource/' + gk_file
                with open(txt_file) as gk:
                    xml=gk.read()
                    soup=BeautifulSoup(xml, 'xml')
                    completeText=''
                    for lines in soup.find_all('l'):
                        completeText += str(lines.string)
                    beta_code_text = bleach.clean(completeText, strip=True).upper()
                    a_replacer = Replacer()
                    unicode_converted = a_replacer.beta_code(beta_code_text)
                    unicode_directory = root + '/cltk_data/greek/text/perseus_unicode/'
                    unic_file_exist = os.path.isdir(unicode_directory)
                    if unic_file_exist is True:
                        pass
                    else:
                        os.mkdir(unicode_directory)
                    author_path = unicode_directory + author
                    author_path_pres = os.path.isdir(author_path)
                    if author_path_pres is True:
                        pass
                    else:
                        os.mkdir(author_path)
                    gk_file_txt = os.path.splitext(gk_file)[0] + '.txt'
                    uni_write = author_path + '/' + gk_file_txt
                    with open(uni_write, 'w') as uni_write:
                        uni_write.write(unicode_converted)
```

```python
def super_organize(text):
    organize_text=[]
    text=tlg_plaintext_cleanup(text, rm_punctuation=True, rm_periods=False)
    text2=lemmatizer.lemmatize(text)
    organize_text.append(text2)
    return organize_text

super_organize('μηδαμῶς, ὦ φίλτατ’ ἀνδρῶν, ἄλλα δράσωμεν κακά. ἀλλὰ καὶ τάδ’ ἐξαμῆσαι πολλά, δύστηνον θέρος πημονῆς δ’ ἅλις γ’ ὑπάρχει· μηδὲν αἱματώμεθα. στείχετ’ αἰδοῖοι γέροντες πρὸς δόμους, πεπρωμένοις [τούσδε] πρὶν παθεῖν εἴξαντες ὥρᾳ· χρῆν τάδ’ ὡς ἐπράξαμεν. εἰ δέ τοι μόχθων γένοιτο τῶνδ’ ἅλις, δεχοίμεθ’ ἄν,')
```

# Mass Cleaning of Files

```python
import os

root = '/Users/wimsey'
unicode_root = root + '/cltk_data/greek/text/perseus_unicode/'
authors = os.listdir(unicode_root)


for author in authors:
    if not '.' in author:
        texts = os.listdir(unicode_root + author)
    for text in texts:
        text_match = re.match(r'.*_gk.txt', text)
        if text_match:
            gk_file = text_match.group()
            txt_file = unicode_root + author + '/' + gk_file
            with open(txt_file) as gk:
                txt=gk.read()
                organize_greek=super_organize(txt)
                clean_greek=super_clean(organize_greek)
                clean_directory=root + '/cltk_data/greek/text/clean_greek/'
                clean_file_exist = os.path.isdir(clean_directory)
                if clean_file_exist is True:
                    pass
                else:
                    os.mkdir(clean_greek)
                author_path = clean_directory + author
                author_path_pres = os.path.isdir(author_path)
                if author_path_pres is True:
                    pass
                else:
                    os.mkdir(author_path)           
```

# Vectors

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
aesch_dir='/Users/wimsey/cltk_data/greek/text/perseus_unicode/Aeschylus/'
listdir(aesch_dir)
filenames=[aesch_dir + 'aesch.ag_gk.txt',
          aesch_dir + 'aesch.eum_gk.txt',
          aesch_dir + 'aesch.lib_gk.txt',
          aesch_dir + 'aesch.pb_gk.txt',
          aesch_dir + 'aesch.pers_gk.txt',
          aesch_dir + 'aesch.seven_gk.txt',
          aesch_dir + 'aesch.supp_gk.txt']
vectorizer=CountVectorizer(input='filename')
dtm=vectorizer.fit_transform(filenames)
vocab=vectorizer.get_feature_names()
```

```python
thumos_indx=list(vocab).index('θυμος')
dtm[0, thumos_index]
```

```python
n, _ = dtm.shape
dist = np.zeros((n, n))
for i in range(n):
     for j in range(n):
         x, y = dtm[i, :], dtm[j, :]
         dist[i, j] = np.sqrt(np.sum((x - y)**2))
from sklearn.metrics.pairwise import euclidean_distances
dist = euclidean_distances(dtm)
np.round(dist, 1)
```
