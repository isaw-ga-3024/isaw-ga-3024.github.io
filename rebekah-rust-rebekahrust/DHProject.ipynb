```python
#The following code is basically from a file found in the cltk_data file.
#on my computer it is found at
#/Users/wimsey/cltk_data/greek/text/greek_text_perseus/perseus_compiler.py
#It's lurking among the author files. It didn't work for me initially and so I made some changes
#to reflect my computer's directory structure.
#I also renamed some variables to be less confusing to me.
import os
import re
import bleach
from cltk.corpus.greek.beta_to_unicode import Replacer

#root = os.getcwd()
root = '/Users/wimsey'
perseus_root = root + '/cltk_data/greek/text/greek_text_perseus/'
#print(perseus_root)
#
authors = os.listdir(perseus_root)

for author in authors:
    if not '.' in author:
        texts = os.listdir(perseus_root + author + '/opensource')
        for text in texts:
            text_match = re.match(r'.*_gk.xml', text)
            if text_match:
                gk_file = text_match.group()
                txt_file = perseus_root + author + '/opensource/' + gk_file
                with open(txt_file) as gk:
                    xml = gk.read()
                    beta_code_text = bleach.clean(xml, strip=True).upper()
                    a_replacer = Replacer()
                    unicode_converted = a_replacer.beta_code(beta_code_text)
                    unicode_directory = root + '/cltk_data/greek/text/perseus_unicode/'
                    #from the folowing code to the next hashtag, why is this inside the loop?
                    #Why not place it before the 'for author in authors' part?
                    unic_file_exist = os.path.isdir(unicode_directory)
                    if unic_file_exist is True:
                        pass
                    else:
                        os.mkdir(unicode_directory)
                    #
                    author_path = unicode_directory + author
                    author_path_pres = os.path.isdir(author_path)
                    if author_path_pres is True:
                        pass
                    else:
                        os.mkdir(author_path)
                    gk_file_txt = os.path.splitext(gk_file)[0] + '.txt'
                    uni_write = author_path + '/' + gk_file_txt
                    with open(uni_write, 'w') as uni_write:
                        uni_write.write(unicode_converted)
```

```python
def list_textfiles(directory):
    textfiles = []
    for filename in listdir(directory):
        if filename.endswith(".txt"):
            textfiles.append(directory + "/" + filename)
    return textfiles
```
